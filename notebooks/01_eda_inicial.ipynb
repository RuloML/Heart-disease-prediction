Te dejamos 6 datasets recomendados para que elijas, de forma que puedas aprender a hacer tu primer proyecto.

- Clasificar emails como spam: https://www.kaggle.com/datasets/abdallahwagih/spam-emails/data

- Predicción del precio de casas en Amsterdam: https://www.kaggle.com/datasets/thomasnibb/amsterdam-house-price-prediction/data

- Predicción de fallo de sensores: https://www.kaggle.com/datasets/umerrtx/machine-failure-prediction-using-sensor-data/data
